```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd
df=pd.read_csv('fetal_health.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df1=df.drop_duplicates()
```

```python
df1.info()
```

separate x and y

```python
x=df1.drop(columns=['fetal_health'])
y=df1[['fetal_health']]
```

```python
x.head()
```

```python
y.head()
```

preprocessing pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
num_pipe=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler',StandardScaler())]).set_output(transform='pandas')
```

```python
x_pre=num_pipe.fit_transform(x)
x_pre.head()
```

train test split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x_pre,y,test_size=0.2,random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

creating gradient boosting model

```python
from sklearn.ensemble import GradientBoostingClassifier
model=GradientBoostingClassifier(learning_rate=0.001,n_estimators=2,max_depth=2,min_samples_split=5)
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

hyperparametric tunning

```python
params={'learning_rate':[0.01,0.05,0.1,0.2],
        'n_estimators':[1,2,3,4,5],
        'max_depth':[1,2,3,4,5,6,7],
        'min_samples_split':[2,3,4,5,6,7,8,9,10]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
gdc=GradientBoostingClassifier()
rscv=RandomizedSearchCV(gdc,param_distributions=params,cv=5,scoring='accuracy')
```

```python
rscv.fit(xtrain,ytrain)
```

```python
rscv.best_params_
```

```python
rscv.best_score_
```

```python
best_gbc=rscv.best_estimator_
best_gbc
```

evaluting the scores


```python
best_gbc.score(xtrain,ytrain)
```

```python
best_gbc.score(xtest,ytest)
```

predicting results

```python
ypred_tr=best_gbc.predict(xtrain)
ypred_ts=best_gbc.predict(xtest)
```

```python
ypred_tr[0:5]
```

```python
ytrain.head()
```

```python
ypred_ts[0:20]
```

```python
ytest.head(20)
```

ckeck confusion matrix

```python
c=best_gbc.classes_
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cf=confusion_matrix(ytest,ypred_ts)
cfd=ConfusionMatrixDisplay(cf,display_labels=c)
cfd.plot()
```

cheeck classification report

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred_ts))
```

predict out of sample data

```python
xnew=pd.read_csv('fetal_health_new.csv')
xnew.head()
```

```python
xnew.info()
```

```python
xnew_pre=num_pipe.transform(xnew)
xnew_pre
```

```python
xnew_pre=pd.DataFrame(xnew_pre)
xnew_pre.head()
```

```python
preds=best_gbc.predict(xnew_pre)
preds
```

```python
probs=best_gbc.predict_proba(xnew_pre)
probs
```

```python
max_prob=probs.max(axis=1)
max_prob
```

```python
xnew['fetal_health']=preds
xnew['max_prob']=max_prob
xnew.head()
```

```python
xnew.to_csv('fetal_health_pred.csv',index=False)
```
