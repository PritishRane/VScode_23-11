```python
import pandas as pd
```

```python
df=pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df.columns=['CustomerID', 'Gender', 'Age', 'Income',
       'Spend']
df.head()
```

```python
x=df[['Age','Income','Spend']]
x.head()
```

preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='median')),('scaler',StandardScaler())]).set_output(transform='pandas')
```

```python
x_pre=num_pipe.fit_transform(x)
x_pre.head()
```

calculate silhouette score

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
n=list(range(2,15))
n
```

```python
scores=[]
for i in n:
    model=AgglomerativeClustering(n_clusters=i)
    model.fit(x_pre)
    # predict the model cluster
    lb=model.labels_
    # calculate silhouette score
    sil=silhouette_score(x_pre,lb)
    scores.append(sil)
```

```python
scores
```

visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))
plt.plot(n,scores)
plt.xlabel('No. of clusters')
plt.ylabel('silhouette score')
plt.title('silhoutte score for verious cluster')
plt.grid()
plt.show()
```

creating agglomerate model

```python
model=AgglomerativeClustering(n_clusters=6)
model.fit(x_pre)
```
